## Nomor 5
Buatlah program yang dapat menampilkan output berupa $10$ bilangan pertama dari barisan Fibonacci 
$$
f_n=f_{n-1}+f_{n-2}.
$$
(Petunjuk: $f_0=f_1=1, f_2=2$ dengan $f_n$ adalah bilangan Fibonacci ke-$n$.)

````python
## Tuliskan jawaban Anda disini
def fibonacci(n):
    if n == 0 or n == 1:
        return 1
    elif n == 2:
        return 2
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

n = fibonacci(10)
print(n)
````

## Nomor 6
Buatlah program yang dapat menampilkan output berupa $5$ jumlahan parsial pertama dari deret 
$$
\sum_{n=0}^\infty \frac{1}{n^2}.
$$

````python
## Tuliskan jawaban Anda disini
jumlah = 0
for i in range(5):
    if i == 0:
        jumlah += 0
    else: 
        jumlah += 1 / (i**2)

print(jumlah)
````

## Nomor 7
Buatlah program yang dapat menampilkan output berupa bentuk bilangan desimal dari bilangan biner
$$
1111111111111000000000000.
$$
Apa fakta menarik yang Anda ketahui tentang bilangan ini?

````python
## Tuliskan jawaban Anda disini
int("1111111111111000000000000", 2) # based on 2 elements (1 and 0)

# (2^13 - 1) * 2^12
````

#### Fakta menarik terkait hal ini yaitu

Pada umumnya ketika kita melakukan konversi angka biner ke suatu angka biasa adalah dengan cara `mengalikan angka dari kanan ke kiri dengan 2^k untuk k = 0, 1, 2, ..., n dan dilakukan penjumlahan untuk masing-masing hasil perkalian`

contoh:
```
angka biner: 101

-> (1 x 2^2) + (0 x 2^1) + (1 x 2^0) 
= 4 + 0 + 1
= 5
```

**akan tetapi** terdapat pola dalam angka biner yang dapat mempermudah dalam melakukan perhitungan ke dalam sebuah angka readable

ketika terdapat angka biner dengan n angka 1 berderet di depan bilangan, kita dapat dengan mudah menghitung dengan rumusan `2^n - 1`

contoh:
```
biner: 1111

n = 4
-> (2^n - 1)
= (2^4 - 1)
= 16 - 1 = 15
```

**selain itu** ketika terdapat n angka 0 di belakang angka biner, kita dapat dengan mudah menghitung dengan mengkalikan angka depan biner dengan (2^n) dengan syarat `angka depan merupakan n angka 1 pada biner`

contoh: 
```
1110000

111  -> n = 3, (2^3 - 1) = 8 - 1 = 7
0000 -> n = 4, (2^4) = 16

1110000 = 7 x 16 = 112
```

akan tetapi ketika terdapat gabungan angka seperti `11101000`, maka angka biner ini tidak termasuk pola yang disebutkan diatas sehingga untuk melakukan konversi ke angka biasa dilakukan dengan cara pada umumnya.

### Jenis-jenis Error dalam Python

- #### **Syntax Error**

Terjadi karena kesalahan penulisan kode

Contoh:

```python
# --- Contoh Salah ---
if 5 > 3
    print("Benar")

# --- Contoh Benar ---
if 5 > 3:
    print("Benar")
```

Program bakalan error jika terdapat _Syntax Error_ seperti berikut:

```
  File "d:\PLM\r.py", line 1
    if 5 > 3
             ^
SyntaxError: expected ':'
```

- #### **Indentation Error**

Python sangat sensitif terhadap indentasi pada program yang dibuat sehingga jika terdapat program yang memiliki kesalahan terdapat indentasi maka akan muncul error _Indentation Error_

Contoh:

```python
# --- Contoh Salah ---
if 5 > 3:
print("Benar")

# --- Contoh Benar ---
if 5 > 3:
    print("Benar")
```

Program bakalan error jika terdapat _Indentation Error_ seperti berikut:

```
  File "d:\PLM\r.py", line 2
    print("Benar")
    ^
IndentationError: expected an indented block
after 'if' statement on line 1
```

- #### **Name Error**

Biasanya terjadi ketika terdapat suatu variable yang dipanggil sebelum variable tersebut dideklarasikan terlebih dahulu

Contoh:

```python
# --- Contoh Salah ---
print(x)

# --- Contoh Benar ---
x = "Benar"
print(x)
```

Program bakalan error jika terdapat _Name Error_ seperti berikut:

```
Traceback (most recent call last):
  File "d:\PLM\r.py", line 1, in <module>
    print(x)
          ^
NameError: name 'x' is not defined
```

- #### **Type Error**

Terjadi ketika tipe data yang tidak sesuai

Contoh:

```python
# --- Contoh Salah ---
print("Umur: " + 20)

# --- Contoh Benar ---
# dilakukan konversi tipe data dari int ke string
print("Umur: " + str(20))
```

Program bakalan error jika terdapat _Type Error_ seperti berikut:

```
Traceback (most recent call last):
  File "d:\PLM\r.py", line 1, in <module>
    print("Umur: " + 20)
          ~~~~~~~~~^~~~
TypeError: can only concatenate str (not "int") to str
```

- #### **Value Error**

Terjadi ketika tipe data benar tetapi nilainya tidak sesuai

Contoh:

```python
# --- Contoh Salah ---
x = int("abc")

# --- Contoh Benar ---
x = int("123") # output: 123 dengan type int
```

Program bakalan error jika terdapat _Value Error_ seperti berikut:

```
Traceback (most recent call last):
  File "d:\PLM\r.py", line 1, in <module>
    x = int('abc')
        ^^^^^^^^^^
ValueError: invalid literal for int() with base 10: 'abc'
```

- #### **Index Error**

Terjadi ketika mengakses indeks list yang tidak ada

Contoh:

```python
# --- Contoh Salah ---
data = [1, 2, 3]
print(data[5])

# --- Contoh Benar ---
data = [1, 2, 3]
print(data[1]) # output: 2
```

Program bakalan error jika terdapat _Index Error_ seperti berikut:

```
Traceback (most recent call last):
  File "d:\PLM\r.py", line 2, in <module>
    print(data[5])
          ~~~~^^^
IndexError: list index out of range
```

- #### **Key Error**

Terjadi ketika mengakses key dictionary yang tidak ada

Contoh:

```python
# --- Contoh Salah ---
data = {"nama": "Budi"}

# key 'umur' tidak  ada pada dict data
print(data["umur"])


# --- Contoh Benar ---
data = {"nama": "Budi"}

print(data["nama"]) # output: Budi
# atau
print(data.get("nama")) # output: Budi
```

Program bakalan error jika terdapat _Key Error_ seperti berikut:

```
Traceback (most recent call last):
  File "d:\PLM\r.py", line 2, in <module>
    print(data["umur"])
          ~~~~^^^^^^^^
KeyError: 'umur'
```

- #### **Zero Division Error**

Terjadi ketika membagi dengan nol

Contoh:

```python
# --- Contoh Salah ---
print(10 / 0)
```

Program bakalan error jika terdapat _Zero Division Error_ seperti berikut:

```
Traceback (most recent call last):
  File "d:\PLM\r.py", line 1, in <module>
    print(10 / 0)
          ~~~^~~
ZeroDivisionError: division by zero
```

### Error Handling

Terkadang ketika membuat suatu program terdapat beberapa error yang muncul pada program dan biasanya dari error tersebut kita tidak mau muncul begitu saja pada output terminal, sehingga
kita bisa melakukan `Error Handling` pada Python menggunakan `try-catch` untuk mengatasi hal tersebut

Struktur dasar `try-catch`:

```python
try:
    # Error code
except JenisError: # -> Jenis error yang mau dihandle
    # Error handler
else:
    # dijalankan jika tidak ada error
finally:
    # Log ketika program berhasil dijalankan
```

Contoh:

```python
try:
    angka = int(input("Masukkan angka: "))
    hasil = 10 / angka
except ValueError:
    print("Input harus angka")
except ZeroDivisionError:
    print("Tidak boleh nol")
else:
    print("Hasil:", hasil)
finally:
    print("Program selesai")
```
