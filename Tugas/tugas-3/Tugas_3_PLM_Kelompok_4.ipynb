## B. Topik Penelitian

- Penelitian ini membahas **perubahan pola pembelian daring produk elektronik konsumen di negara-negara Eropa** sebelum dan selama pandemi COVID-19.
- Analisis dilakukan dengan mempertimbangkan **perbedaan berdasarkan generasi (Gen X dan Gen Y) serta gender**.

## C. Latar Belakang

- Perkembangan digitalisasi di Eropa masih menunjukkan **kesenjangan antarnegara** (_digital divide_).
- Pandemi COVID-19 mempercepat penggunaan teknologi digital, termasuk dalam aktivitas **e-commerce**.
- Oleh karena itu, penelitian ini dilakukan untuk melihat **apakah pandemi memengaruhi perilaku pembelian online** dan **apakah perubahan tersebut terjadi secara merata di seluruh negara Eropa**.

## D. Tujuan Penelitian

- Menganalisis dampak pandemi COVID-19 terhadap **pola pembelian daring produk elektronik konsumen** di negara-negara Eropa.
- Mengidentifikasi **perbedaan pola pembelian** berdasarkan **generasi** dan **gender**.

## E. Data/Objek yang Digunakan

- Penelitian ini menggunakan **data sekunder** dari **Eurostat**, yaitu lembaga statistik resmi Uni Eropa.
- Objek penelitian mencakup **35 negara di Eropa**.
- Data yang dianalisis membandingkan dua periode, yaitu:
  - **2018** sebagai kondisi **sebelum pandemi**, dan
  - **2021** sebagai kondisi **selama/akibat pandemi COVID-19**.
- Variabel yang digunakan adalah **persentase masyarakat yang membeli produk elektronik secara online**.
- Data dibedakan berdasarkan beberapa kelompok, yaitu:
  - seluruh individu,
  - **Generasi Y (Gen Y)**,
  - **Generasi X (Gen X)**,
  - serta **laki-laki dan perempuan** dalam masing-masing kelompok generasi.
- Pembagian ini memungkinkan analisis perilaku pembelian online secara lebih rinci berdasarkan **negara, usia, dan gender**.

## F. Metode Secara Umum

- Penelitian ini menggunakan pendekatan **kuantitatif** dengan analisis berbasis **clustering**.
- Tujuannya adalah untuk melihat **kemiripan pola pembelian online antarnegara**.
- Tahapan penelitian secara umum meliputi:
  - pengumpulan data dari Eurostat,
  - analisis statistik deskriptif,
  - pengujian distribusi data dengan **Shapiro–Wilk test**,
  - penentuan jumlah cluster menggunakan **elbow method**,
  - pengelompokan data dengan **Fuzzy C-Means Clustering**,
  - visualisasi hasil menggunakan **t-SNE**,
  - serta perbandingan hasil cluster antara tahun 2018 dan 2021.
- Metode **Fuzzy C-Means** dipilih karena memungkinkan suatu negara memiliki **tingkat kedekatan dengan lebih dari satu cluster**, sehingga lebih sesuai untuk data yang bersifat beragam dan tidak tegas.

## G. Hasil Utama

- Hasil penelitian menunjukkan bahwa **pandemi COVID-19 memengaruhi pola pembelian online produk elektronik** di negara-negara Eropa.
- Namun, perubahan tersebut **tidak terjadi secara merata** di semua negara.
- Berdasarkan hasil clustering:
  - pada **tahun 2018** terbentuk **3 cluster**,
  - sedangkan pada **tahun 2021** terbentuk **4 cluster**.
- Bertambahnya jumlah cluster menunjukkan bahwa setelah pandemi, **pola pembelian online antarnegara menjadi lebih beragam**.
- Penelitian ini juga menemukan adanya **perbedaan kecenderungan perilaku pembelian** berdasarkan **generasi** dan **gender**.
- Secara umum, hasil ini menunjukkan bahwa pandemi mempercepat perkembangan **e-commerce**, sekaligus memperlihatkan **perbedaan tingkat adaptasi digital** di berbagai negara Eropa.

## H. Kesimpulan Umum

- Penelitian ini menyimpulkan bahwa **pandemi COVID-19 memberikan dampak terhadap perilaku pembelian online produk elektronik** di negara-negara Eropa.
- Perubahan yang terjadi **tidak bersifat homogen**, karena setiap negara memiliki **tingkat adaptasi digital yang berbeda**.
- Melalui metode **Fuzzy C-Means Clustering**, penelitian ini menunjukkan adanya **pergeseran pola perilaku belanja online** dari sebelum pandemi ke masa pandemi.
- Dengan demikian, perkembangan **e-commerce** selama pandemi tidak hanya menunjukkan peningkatan aktivitas belanja online, tetapi juga memperlihatkan **variasi perilaku konsumen antarnegara dan antarkelompok demografis**.

# **III. Landasan Teori**


Penelitian oleh Jajić et al. (2025) ini bertujuan untuk menganalisis bagaimana pandemi COVID-19 mengubah pola belanja online produk elektronik konsumen di 35 negara Eropa, dengan fokus khusus pada perbedaan antara Generasi X dan Generasi Y.

**1. Kesenjangan Digital (_Digital Divide_)**

> Landasan utama penelitian ini adalah konsep kesenjangan digital, yaitu perbedaan antara kelompok demografis atau wilayah dalam hal akses dan penggunaan teknologi informasi modern. Proses ini biasanya dimulai dari tingkat makro (negara) sebelum menyebar ke tingkat mikro (perusahaan dan individu). Meskipun Komisi Eropa telah meluncurkan indeks seperti Digital Economy and Society Index (DESI) untuk menyelaraskan perkembangan digital, ketimpangan antarnegara masih tetap ada. COVID-19 secara tidak langsung membantu memperkecil kesenjangan ini karena memaksa masyarakat menggunakan platform e-commerce untuk kebutuhan sehari-hari.

**2. _E-commerce_ dan Pengaruh COVID-19**

> Teori ini menjelaskan transisi perilaku ekonomi dari toko tradisional ke transaksi elektronik. Pandemi memaksa bisnis mengadopsi penjualan daring dan kerja jarak jauh demi mematuhi protokol pembatasan sosial. Kecepatan transisi digital suatu negara sangat dipengaruhi oleh tingkat kemajuan ekonomi nasional tersebut. Penjualan daring terbukti meningkatkan total penjualan dan mengalihkan konsumen dari saluran tradisional, meskipun dampak terhadap dispersi harga internasional dianggap tidak terlalu signifikan dalam penelitian terdahulu.

**3. Teori Kohort Generasi (Generasi X dan Y)**

> Paper ini menggunakan teori generasi untuk melihat perbedaan perilaku konsumsi berdasarkan usia. Berbagai generasi memiliki cara berbeda dalam memandang risiko saat berbelanja secara tradisional maupun daring.

> **a) Generasi Y (Millennials):** Kelompok ini menunjukkan lonjakan perilaku belanja online yang signifikan selama pandemi, meskipun secara statistik rata-rata mereka mengalami penurunan persentase pembelian pada tahun 2021 dibanding 2018 (mungkin karena sudah membeli peralatan di awal pandemi).

> **b) Generasi X:** Dibandingkan Gen Y, kelompok ini menunjukkan keterlibatan yang lebih stabil dan konsisten dalam rutinitas belanja online mereka tanpa peningkatan yang signifikan selama pandemi.

**4. _Fuzzy C-Means_ (FCM) _Clustering_**

> Fuzzy C-Means (FCM) adalah algoritma pengelompokan fuzi (soft clustering) populer dalam penambangan data yang memungkinkan setiap objek menjadi anggota dari beberapa klaster secara bersamaan dengan derajat keanggotaan antara 0 hingga 1.

> Secara teoretis, FCM bekerja dengan meminimalkan fungsi objektif berikut:

$$J_m = \sum_{i=1}^{N} \sum_{j=1}^{C} u_{ij}^m \|x_i - v_j\|^2$$

> Di mana $u_{ij}$ adalah derajat keanggotaan data $i$ pada klaster $j$, dan $v_j$ adalah pusat klaster. Berbeda dengan metode tegas (hard clustering), FCM bekerja dengan meminimalkan fungsi objektif untuk mengurangi ketidaksamaan dalam satu kelompok dan meningkatkan kemiripan antar-anggota. Meskipun efektif, algoritma ini sangat sensitif terhadap inisialisasi pusat klaster awal yang dapat memengaruhi kecepatan konvergensi. Konsep ini digunakan dalam paper Jajić et al. (2025) untuk mengelompokkan 35 negara Eropa berdasarkan kemiripan pola belanja warga negaranya, sehingga memungkinkan analisis yang lebih halus terhadap pola laten dan diskrepansi regional yang tidak dapat ditangkap oleh metode pengelompokan tradisional.

**5. Reduksi Dimensi dan Visualisasi (t-SNE)**

> Penelitian ini menggunakan algoritma t-Distributed Stochastic Neighbour Embedding (t-SNE) sebagai kerangka kerja untuk memvisualisasikan data yang kompleks. t-SNE adalah teknik pembelajaran mesin (machine learning) yang digunakan untuk reduksi dimensi non-linear. Tujuannya adalah menempatkan poin-poin data dari ruang berdimensi tinggi ke dalam grafik dua atau tiga dimensi. Algoritma ini bekerja dengan merangkum struktur lokal dari data berdimensi tinggi. Artinya, negara-negara yang memiliki pola belanja serupa di dunia nyata akan tetap digambarkan berdekatan satu sama lain dalam grafik. Metodologi ini menyederhanakan proses optimasi sehingga menghasilkan visualisasi yang lebih baik dan akurat dibandingkan metode linier tradisional. Dalam paper ini, t-SNE digunakan untuk membuktikan bahwa solusi klaster yang dihasilkan oleh Fuzzy C-means adalah valid, karena anggota klaster digambarkan berada dalam posisi yang berdekatan secara visual.


# **IV. Metodologi**


## 1. Pengumpulan Data

- Data diambil dari **Eurostat** (badan statistik resmi Uni Eropa).
- Objek penelitian: 35 negara Eropa.
- Variabel: persentase penduduk yang membeli elektronik secara online, dibedakan menurut:
  - Generasi X
  - Generasi Y
  - Laki-laki
  - Perempuan
- Periode: **2018 (pra-pandemi)** vs **2021 (masa pandemi)**.

## 2. Analisis Deskriptif

- Digunakan untuk **menggambarkan data** (rata-rata, minimum, maksimum, standar deviasi).
- **Rumus rata-rata (mean):**
  $$\bar{x} = \frac{\sum x_i}{n}$$
- **Rumus standar deviasi (SD):**
  $$\sigma = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n}}$$
- Contoh: Jika 5 negara punya persentase belanja [10, 20, 30, 40, 50], maka rata-rata = 30%.

## 3. Uji Normalitas (Shapiro-Wilk Test)

- Tujuan: mengecek apakah data mengikuti distribusi normal (bentuk lonceng).
- **Hasil tes:** berupa _p-value_.
  - **p-value** adalah angka probabilitas.
  - Jika p-value > 0,05 → data dianggap normal.
  - Jika p-value < 0,05 → data tidak normal.

#### Rumus Shapiro-Wilk

Statistik uji Shapiro-Wilk ditulis sebagai:

$$
W = \frac{\left(\sum_{i=1}^{n} a_i x_{(i)}\right)^2}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$$

- $x_{(i)}$ = data yang sudah diurutkan (order statistics).
- $\bar{x}$ = rata-rata data.
- $a_i$ = konstanta yang dihitung berdasarkan kovarians data normal.
- $n$ = jumlah data.

## 4. Fuzzy C-Means Clustering

- **Clustering** adalah teknik untuk mengelompokkan data berdasarkan kemiripan.
- **K-Means Clustering**: setiap data hanya masuk ke **satu cluster** (tegas).
- **Fuzzy C-Means Clustering (FCM)**: setiap data bisa masuk ke **lebih dari satu cluster** dengan **derajat keanggotaan** (nilai antara 0–1).
  - Artinya, sebuah negara bisa mirip dengan dua kelompok sekaligus, tapi dengan tingkat kemiripan berbeda.

### Rumus Fungsi Objektif

FCM berusaha meminimalkan fungsi berikut:

$$
J_m = \sum_{i=1}^{N} \sum_{j=1}^{C} u_{ij}^m \cdot \|x_i - c_j\|^2
$$

- $N$ = jumlah data (misalnya jumlah negara)
- $C$ = jumlah cluster
- $u_{ij}$ = derajat keanggotaan data ke-$i$ pada cluster ke-$j$ (antara 0–1)
- $m$ = parameter fuzziness (biasanya $m=2$)
- $x_i$ = data ke-$i$ (misalnya persentase belanja online suatu negara)
- $c_j$ = pusat cluster ke-$j$

## 5. Elbow Method

- **Elbow Method** adalah teknik untuk menentukan jumlah cluster optimal dalam analisis clustering.
- Disebut "elbow" (siku) karena grafik hasil perhitungan biasanya membentuk sudut seperti siku tangan.
- Dalam clustering, kita ingin setiap data berada dekat dengan pusat cluster.
- Ukuran kedekatan ini disebut **Within Sum of Squares (WSS)**:
  $$
  WSS = \sum_{i=1}^{N} \|x_i - c_{cluster(i)}\|^2
  $$
  - $N$ = jumlah data
  - $x_i$ = data ke-$i$
  - $c_{cluster(i)}$ = pusat cluster tempat data $i$ berada
- Semakin kecil WSS → semakin rapat data dalam cluster.

## 6. Validasi Model

Digunakan untuk memastikan cluster yang terbentuk memang masuk akal:

- **R² (Explained Variance):**
  $$R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$$
  → Semakin mendekati 1, semakin baik.

- **AIC (Akaike Information Criterion):**
  $$AIC = 2k - 2\ln(L)$$
  → Semakin kecil AIC, semakin baik.

- **BIC (Bayesian Information Criterion):**
  $$BIC = k \ln(n) - 2\ln(L)$$
  → Lebih ketat daripada AIC, menghukum model yang terlalu kompleks.

- **Silhouette Score:**
  $$s = \frac{b-a}{\max(a,b)}$$
  - $a$ = rata-rata jarak ke anggota cluster sendiri
  - $b$ = rata-rata jarak ke cluster terdekat
    → Nilai mendekati 1 = cluster jelas, mendekati 0 = cluster tumpang tindih.

## 7. Visualisasi dengan t-SNE

- **t-SNE (t-distributed Stochastic Neighbor Embedding)** adalah teknik visualisasi data.
- Digunakan untuk **menyederhanakan data berdimensi tinggi** (misalnya banyak variabel: Gen X, Gen Y, gender, tahun) ke dalam **grafik 2D atau 3D**.
- Tujuannya: agar pola cluster lebih mudah dilihat oleh manusia.

- t-SNE bekerja dengan prinsip **tetangga dekat**:
  - Jika dua negara punya pola belanja online mirip, maka titik mereka akan ditempatkan berdekatan di grafik.
  - Jika pola belanja berbeda jauh, titik mereka akan ditempatkan berjauhan.
- Dengan cara ini, cluster yang terbentuk bisa dilihat secara visual.

#### Rumus Inti (Konsep Probabilitas)

1. Hitung **probabilitas kedekatan** antar data di dimensi tinggi:

   $$
   p_{ij} = \frac{\exp(-\|x_i - x_j\|^2 / 2\sigma^2)}{\sum_{k \neq l} \exp(-\|x_k - x_l\|^2 / 2\sigma^2)}
   $$
   - $x_i, x_j$ = data ke-$i$ dan ke-$j$
   - $\sigma$ = parameter skala jarak
   - $p_{ij}$ = probabilitas bahwa $x_i$ dan $x_j$ adalah tetangga dekat

2. Hitung probabilitas kedekatan di dimensi rendah (2D/3D):

   $$
   q_{ij} = \frac{(1 + \|y_i - y_j\|^2)^{-1}}{\sum_{k \neq l} (1 + \|y_k - y_l\|^2)^{-1}}
   $$
   - $y_i, y_j$ = posisi data ke-$i$ dan ke-$j$ di ruang 2D/3D
   - $q_{ij}$ = probabilitas kedekatan di ruang visualisasi

3. t-SNE berusaha meminimalkan perbedaan antara $p_{ij}$ dan $q_{ij}$ dengan menggunakan **Kullback-Leibler Divergence (KL Divergence)**:
   $$
   KL(P||Q) = \sum_{i \neq j} p_{ij} \log \frac{p_{ij}}{q_{ij}}
   $$

---

### B. Preprocessing Data

> **Deskripsi Tahap 2**  
> Preprocessing bertujuan membersihkan dan mentransformasi data mentah agar siap digunakan oleh algoritma **Fuzzy C-Means**.  
> Dua sub-tahap dijalankan secara berurutan:
>
> 1. Penanganan _missing values_
> 2. Deteksi dan koreksi _outlier_

---

### C. Normalisasi Data

> **Deskripsi Tahap 3**  
> Normalisasi dilakukan menggunakan metode **Min-Max Normalization** untuk menyeragamkan skala seluruh fitur ke rentang [0, 1].  
> Hal ini penting agar fitur dengan nilai besar tidak mendominasi perhitungan jarak pada algoritma FCM.
>
> **Formula:**
> $$x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$$
>
> Normalisasi diterapkan per kolom (per fitur), sehingga setiap fitur memiliki nilai minimum 0 dan maksimum 1 secara independen.

---

### D. Inisialisasi Parameter FCM

> **Deskripsi Tahap 4**  
> Sebelum algoritma dijalankan, parameter-parameter berikut ditetapkan:
>
> | Parameter | Nilai               | Keterangan                                            |
> | --------- | ------------------- | ----------------------------------------------------- |
> | C         | 3 (2018) / 4 (2021) | Jumlah cluster, ditentukan dari elbow method          |
> | m         | 2                   | Fuzzifier, mengontrol tingkat kekaburan antar cluster |
> | ε         | 0.0001              | Toleransi konvergensi                                 |
> | max_iter  | 100                 | Batas maksimum iterasi                                |
> | seed      | 42                  | Seed random untuk reproduktibilitas                   |
>
> Matriks keanggotaan awal **U** diinisialisasi secara acak dengan syarat setiap baris berjumlah 1, merepresentasikan distribusi keanggotaan tiap negara terhadap seluruh cluster.

---

### E. Penentuan Jumlah Cluster (Elbow Method)

> **Deskripsi Tahap 5**  
> Jumlah cluster optimal ditentukan menggunakan **Elbow Method** dengan memplot nilai _Objective Function_ (J) terhadap berbagai nilai C.  
> Titik "siku" (_elbow_) pada grafik menunjukkan nilai C di mana penambahan cluster tidak lagi memberikan penurunan J yang signifikan.
>
> **Objective Function FCM:**
> $$J = \sum_{i=1}^{C} \sum_{j=1}^{N} u_{ij}^m \cdot d_{ij}^2$$
>
> Di mana $d_{ij}$ adalah jarak Euclidean antara data ke-$j$ dengan centroid cluster ke-$i$.

---

### F. Proses Clustering FCM

> **Deskripsi Tahap 6**  
> Algoritma **Fuzzy C-Means** dijalankan secara iteratif hingga konvergen melalui langkah-langkah berikut:
>
> **Langkah 1 — Hitung Centroid:**
> $$V_i = \frac{\sum_{j=1}^{N} u_{ij}^m \cdot x_j}{\sum_{j=1}^{N} u_{ij}^m}$$
>
> **Langkah 2 — Update Membership:**
> $$u_{ij} = \frac{1}{\sum_{k=1}^{C} \left(\frac{d_{ij}}{d_{kj}}\right)^{\frac{2}{m-1}}}$$
>
> **Langkah 3 — Cek Konvergensi:**  
> Iterasi berhenti jika $\|U_{baru} - U_{lama}\| \leq \varepsilon$ atau iterasi mencapai batas maksimum.
>
> Setiap negara memiliki derajat keanggotaan di semua cluster secara bersamaan (bukan hard assignment), dan label akhir ditentukan dari nilai keanggotaan tertinggi.

---

### G. Visualisasi Hasil

> **Deskripsi Tahap 7**  
> Hasil clustering divisualisasikan menggunakan **t-SNE** (_t-Distributed Stochastic Neighbor Embedding_) untuk mereduksi 14 dimensi data menjadi 2 dimensi yang dapat diplot.  
> t-SNE dipilih karena mampu mempertahankan struktur lokal data berdimensi tinggi dengan lebih baik dibandingkan PCA, sesuai dengan metode yang digunakan pada paper acuan.
>
> Setiap titik pada scatter plot merepresentasikan satu negara, diwarnai berdasarkan cluster hasil FCM, disertai label nama negara untuk memudahkan interpretasi.

---

### H. Analisis dan Interpretasi

> **Deskripsi Tahap 8**  
> Hasil clustering 2018 dan 2021 dibandingkan untuk mengidentifikasi:
>
> 1. **Perubahan jumlah cluster** — apakah pola belanja makin beragam pasca pandemi
> 2. **Perpindahan negara antar cluster** — negara mana yang mengalami peningkatan atau penurunan perilaku belanja online
> 3. **Konsistensi cluster** — negara yang tetap berada di cluster yang sama di kedua tahun
>
> Interpretasi mengacu pada konteks paper: negara dengan nilai persentase tinggi cenderung masuk cluster "advanced e-commerce", sementara negara dengan nilai rendah masuk cluster "lagging".

```python
"""
=============================================================
  PREPROCESSING PIPELINE UNTUK FUZZY C-MEANS CLUSTERING
  Studi: COVID-19 & E-Commerce Eropa (Gen X & Gen Y)
  Data: dummy (5 negara, 14 fitur)
=============================================================
"""

import pandas as pd
import numpy as np
from scipy import stats
import os


# ─── WARNA TERMINAL ───────────────────────────────────────────
class C:
    RESET   = "\033[0m"
    BOLD    = "\033[1m"
    CYAN    = "\033[96m"
    GREEN   = "\033[92m"
    YELLOW  = "\033[93m"
    RED     = "\033[91m"
    BLUE    = "\033[94m"
    MAGENTA = "\033[95m"
    WHITE   = "\033[97m"
    DIM     = "\033[2m"

def header(text):
    bar = "═" * 62
    print(f"\n{C.CYAN}{C.BOLD}{bar}")
    print(f"  {text}")
    print(f"{bar}{C.RESET}\n")

def section(text):
    print(f"\n{C.BLUE}{C.BOLD}{'─'*50}")
    print(f"  {text}")
    print(f"{'─'*50}{C.RESET}")

def info(text):
    print(f"  {C.WHITE}ℹ  {text}{C.RESET}")

def success(text):
    print(f"  {C.GREEN}✔  {text}{C.RESET}")

def warning(text):
    print(f"  {C.YELLOW}⚠  {text}{C.RESET}")

def error(text):
    print(f"  {C.RED}✘  {text}{C.RESET}")

def prompt(text):
    return input(f"\n  {C.MAGENTA}{C.BOLD}▶  {text}{C.RESET} ")

def show_table(df, title=""):
    if title:
        print(f"\n  {C.DIM}{title}{C.RESET}")
    print()
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', 130)
    pd.set_option('display.float_format', lambda x: f'{x:.4f}')
    lines = df.to_string(index=True).split('\n')
    for line in lines:
        print(f"    {C.DIM}{line}{C.RESET}")
    print()


# ═══════════════════════════════════════════════════════════════
# 1. DATA DUMMY (HARDCODED)
# ═══════════════════════════════════════════════════════════════

def load_data():
    header("STEP 1 — DATA DUMMY E-COMMERCE EROPA")

    # Masalah data kotor yang disengaja:
    #   Poland  · All_Individuals_2021 = None   → missing value
    #   Poland  · Women_GenY_2018      = None   → missing value (N/A)
    #   France  · GenY_2021            = 999    → outlier ekstrem
    #   Romania · Women_GenX_2018      = 0      → nilai nol (potensi missing)
    raw = {
        'Country':              ['Germany', 'Poland', 'Romania', 'France', 'Netherlands'],
        'All_Individuals_2018': [24,        8,        3,         18,       27           ],
        'All_Individuals_2021': [22,        None,     2,         17,       25           ],
        'GenY_2018':            [35,        12,       5,         28,       37           ],
        'GenY_2021':            [31,        10,       4,         999,      33           ],
        'GenX_2018':            [20,        7,        2,         15,       22           ],
        'GenX_2021':            [21,        8,        3,         16,       23           ],
        'Men_GenY_2018':        [38,        14,       6,         32,       41           ],
        'Men_GenY_2021':        [34,        11,       5,         28,       37           ],
        'Men_GenX_2018':        [14,        5,        1,         10,       16           ],
        'Men_GenX_2021':        [15,        6,        2,         11,       17           ],
        'Women_GenY_2018':      [18,        None,     4,         13,       20           ],
        'Women_GenY_2021':      [17,        8,        3,         14,       19           ],
        'Women_GenX_2018':      [7,         2,        0,         5,        9            ],
        'Women_GenX_2021':      [8,         3,        1,         6,        10           ],
    }

    df = pd.DataFrame(raw)

    # ── FIX: konversi semua kolom numerik ke float64 sejak awal ──
    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    df[num_cols] = df[num_cols].astype(float)

    info(f"Dataset      : European E-Commerce · COVID-19 Study")
    info(f"Sumber       : Eurostat (data dummy simulasi)")
    info(f"Ukuran data  : {df.shape[0]} baris x {df.shape[1]} kolom")
    info(f"Fitur num.   : {df.shape[1] - 1} kolom (selain Country)")

    section("Catatan Data Kotor yang Disengaja")
    warning("Poland      . All_Individuals_2021 = None  --> missing value")
    warning("Poland      . Women_GenY_2018      = None  --> missing value (N/A)")
    warning("France      . GenY_2021            = 999   --> outlier ekstrem")
    warning("Romania     . Women_GenX_2018      = 0     --> nilai nol (potensi missing)")

    show_table(df, "Raw Data (sebelum preprocessing):")
    return df


# ═══════════════════════════════════════════════════════════════
# 2. PILIHAN PREPROCESSING
# ═══════════════════════════════════════════════════════════════

def ask_preprocessing(df):
    header("STEP 2 — PILIHAN PREPROCESSING")

    num_cols  = df.select_dtypes(include=[np.number]).columns.tolist()
    n_missing = df.isnull().sum().sum()

    info(f"Jumlah baris       : {len(df)}")
    info(f"Kolom numerik      : {len(num_cols)}")
    info(f"Total missing value: {n_missing}")
    print()
    print(f"  {C.WHITE}Apakah Anda ingin melakukan preprocessing pada data ini?{C.RESET}\n")
    print(f"  {C.CYAN}[1]{C.RESET} Ya  — pipeline: Missing Values -> Outlier")
    print(f"  {C.CYAN}[2]{C.RESET} Tidak — gunakan data mentah langsung\n")

    while True:
        choice = prompt("Pilihan Anda [1/2]:").strip()
        if choice == '1':
            success("Preprocessing akan dijalankan.")
            return True
        elif choice == '2':
            warning("Preprocessing dilewati.")
            return False
        else:
            error("Input tidak valid. Masukkan 1 atau 2.")


# ═══════════════════════════════════════════════════════════════
# 3. HANDLE MISSING VALUES
# ═══════════════════════════════════════════════════════════════

def handle_missing(df):
    header("STEP 3 — HANDLE MISSING VALUES")

    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

    # Deteksi & laporan
    section("Laporan Missing Values")
    total_missing = 0
    missing_info  = {}

    for col in df.columns:
        n = df[col].isnull().sum()
        missing_info[col] = n
        if n > 0:
            total_missing += n
            pct    = n / len(df) * 100
            rows   = df[df[col].isnull()].index.tolist()
            negara = df.loc[rows, 'Country'].tolist() if 'Country' in df.columns else rows
            warning(f"Kolom '{col}': {n} missing ({pct:.1f}% dari data)")
            info(f"  --> Negara  : {negara}")
            info(f"  --> Index   : {rows}")

    if total_missing == 0:
        success("Tidak ditemukan missing value pada data.")
    else:
        print(f"\n  {C.BOLD}{C.YELLOW}Total missing value ditemukan: {total_missing}{C.RESET}")

    # Pilih strategi
    section("Pilih Strategi Handling Missing Values")
    print(f"  {C.CYAN}[1]{C.RESET} Imputasi Median")
    print(f"      --> Ganti missing dengan nilai median per kolom")
    print(f"      --> Robust terhadap outlier, cocok untuk data kecil\n")
    print(f"  {C.CYAN}[2]{C.RESET} Imputasi Mean")
    print(f"      --> Ganti missing dengan nilai rata-rata per kolom")
    print(f"      --> Sensitif terhadap outlier\n")
    print(f"  {C.CYAN}[3]{C.RESET} Drop Rows")
    print(f"      --> Hapus baris yang mengandung missing value")
    print(f"      --> Data lebih bersih tapi jumlah baris berkurang\n")

    while True:
        choice = prompt("Pilihan strategi [1/2/3]:").strip()
        if choice in ['1', '2', '3']:
            break
        error("Input tidak valid. Masukkan 1, 2, atau 3.")

    df_result = df.copy()

    if choice == '1':
        strategy = "Imputasi Median"
        for col in num_cols:
            n = df[col].isnull().sum()
            if n > 0:
                med = df[col].median()
                df_result[col] = df[col].fillna(med)
                success(f"Kolom '{col}': {n} missing --> diisi median = {med:.2f}")

    elif choice == '2':
        strategy = "Imputasi Mean"
        for col in num_cols:
            n = df[col].isnull().sum()
            if n > 0:
                avg = df[col].mean()
                df_result[col] = df[col].fillna(avg)
                success(f"Kolom '{col}': {n} missing --> diisi mean = {avg:.2f}")

    else:
        strategy = "Drop Rows"
        before = len(df_result)
        df_result = df_result.dropna()
        dropped = before - len(df_result)
        if dropped > 0:
            warning(f"{dropped} baris dihapus karena mengandung missing value.")
            info(f"Baris tersisa: {len(df_result)} dari {before}")
        else:
            success("Tidak ada baris yang perlu dihapus.")

    # Pastikan kolom numerik tetap float64 setelah operasi
    num_cols_result = df_result.select_dtypes(include=[np.number]).columns.tolist()
    df_result[num_cols_result] = df_result[num_cols_result].astype(float)

    section("Ringkasan Handling Missing Values")
    info(f"Strategi yang dipilih : {C.CYAN}{strategy}{C.RESET}")
    info(f"Total missing (awal)  : {C.YELLOW}{total_missing}{C.RESET}")
    info(f"Missing tersisa       : {C.GREEN}{df_result.isnull().sum().sum()}{C.RESET}")
    info(f"Jumlah baris sesudah  : {len(df_result)}")

    show_table(df_result, "Data setelah handling missing values:")

    print(f"\n  {C.GREEN}{C.BOLD}PENJELASAN:{C.RESET}")
    if choice == '1':
        print(f"  {C.WHITE}Imputasi median dipilih karena robust terhadap outlier dibanding mean.")
        print(f"  Pada dataset kecil (n=5 negara), menghapus baris akan sangat mengurangi")
        print(f"  representasi data. Median tidak terpengaruh nilai ekstrem seperti 999")
        print(f"  pada kolom GenY_2021, sehingga centroid Fuzzy C-means tidak bias.{C.RESET}")
    elif choice == '2':
        print(f"  {C.WHITE}Imputasi mean mengisi missing value dengan rata-rata kolom.")
        print(f"  Perhatikan: outlier (nilai 999 pada GenY_2021) dapat mendistorsi mean")
        print(f"  secara signifikan. Pastikan outlier sudah ditangani sebelum menggunakan")
        print(f"  mean sebagai nilai imputasi pada dataset ini.{C.RESET}")
    else:
        print(f"  {C.WHITE}Drop rows memastikan hanya data lengkap yang digunakan.")
        print(f"  Namun pada dataset kecil (5 negara), setiap baris yang hilang")
        print(f"  berdampak besar (20% per baris). Pertimbangkan imputasi jika")
        print(f"  sampel terlalu kecil untuk clustering yang representatif.{C.RESET}")

    return df_result


# ═══════════════════════════════════════════════════════════════
# 4. DETEKSI & KOREKSI OUTLIER
# ═══════════════════════════════════════════════════════════════

def handle_outlier(df):
    header("STEP 4 — DETEKSI & KOREKSI OUTLIER")

    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

    # ── FIX: pastikan semua kolom numerik bertipe float64
    #    agar capping (hasil float) bisa disimpan tanpa TypeError
    df = df.copy()
    df[num_cols] = df[num_cols].astype(float)

    section("Pilih Metode Deteksi Outlier")
    print(f"  {C.CYAN}[1]{C.RESET} Z-Score  (threshold = 2.5 sigma)")
    print(f"      --> Outlier jika |z-score| > 2.5")
    print(f"      --> Cocok untuk data mendekati distribusi normal\n")
    print(f"  {C.CYAN}[2]{C.RESET} IQR Method  (faktor = 1.5 x IQR)")
    print(f"      --> Outlier jika di luar [Q1 - 1.5xIQR, Q3 + 1.5xIQR]")
    print(f"      --> Lebih robust, cocok untuk distribusi tidak normal\n")

    while True:
        choice = prompt("Pilihan metode [1/2]:").strip()
        if choice in ['1', '2']:
            break
        error("Input tidak valid. Masukkan 1 atau 2.")

    method_name = "Z-Score (2.5 sigma)" if choice == '1' else "IQR (1.5x)"

    section(f"Laporan Deteksi Outlier — Metode {method_name}")
    df_result   = df.copy()
    total_out   = 0
    outlier_log = []

    for col in num_cols:
        vals = df[col].dropna()
        if len(vals) < 2:
            continue

        if choice == '1':
            z       = np.abs(stats.zscore(vals))
            out_idx = vals[z > 2.5].index
            m, s    = vals.mean(), vals.std()
            lo, hi  = m - 2.5 * s, m + 2.5 * s
        else:
            Q1, Q3  = vals.quantile(0.25), vals.quantile(0.75)
            IQR     = Q3 - Q1
            lo, hi  = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
            out_idx = vals[(vals < lo) | (vals > hi)].index

        for idx in out_idx:
            orig   = df_result.loc[idx, col]
            cap    = float(np.clip(orig, lo, hi))
            negara = df_result.loc[idx, 'Country'] if 'Country' in df_result.columns else idx

            warning(f"OUTLIER -- Negara: '{negara}' | Kolom: '{col}'")
            info(f"  Nilai asli        : {C.RED}{orig}{C.RESET}")
            info(f"  Batas [lo, hi]    : [{lo:.2f}, {hi:.2f}]")
            info(f"  Nilai setelah cap : {C.GREEN}{round(cap, 4)}{C.RESET}")

            df_result.loc[idx, col] = round(cap, 4)
            total_out += 1
            outlier_log.append({
                'Negara'    : negara,
                'Kolom'     : col,
                'Nilai Asli': orig,
                'Batas Lo'  : round(lo,  2),
                'Batas Hi'  : round(hi,  2),
                'Nilai Cap' : round(cap, 4),
            })

    section("Ringkasan Koreksi Outlier")
    info(f"Metode           : {C.CYAN}{method_name}{C.RESET}")
    info(f"Kolom diperiksa  : {len(num_cols)}")
    info(f"Total outlier    : {C.YELLOW}{total_out}{C.RESET}")
    info(f"Strategi koreksi : Winsorization (capping ke batas deteksi)")

    if total_out == 0:
        success("Tidak ada outlier yang ditemukan pada data.")
    else:
        log_df = pd.DataFrame(outlier_log)
        show_table(log_df, "Detail outlier yang dikoreksi:")

    show_table(df_result, "Data setelah koreksi outlier:")

    print(f"\n  {C.GREEN}{C.BOLD}PENJELASAN:{C.RESET}")
    print(f"  {C.WHITE}Outlier berbahaya untuk Fuzzy C-means karena algoritma ini")
    print(f"  menghitung derajat keanggotaan berdasarkan jarak Euclidean ke centroid.")
    print(f"  Nilai ekstrem seperti 999 pada GenY_2021 (France) menarik centroid")
    print(f"  jauh dari posisi representatif, menghasilkan cluster yang tidak akurat.")
    print(f"  Winsorization mempertahankan jumlah data sambil membatasi pengaruh")
    print(f"  nilai ekstrem pada perhitungan centroid Fuzzy C-means.{C.RESET}")

    return df_result
# ═══════════════════════════════════════════════════════════════
# 6. SIMPAN HASIL
# ═══════════════════════════════════════════════════════════════

def save_result(df):
    header("STEP 6 — SIMPAN HASIL")

    default_name = "ecommerce_preprocessed.csv"
    print(f"  {C.WHITE}File hasil preprocessing akan disimpan sebagai CSV.{C.RESET}")

    out_path = prompt(f"Nama file output (Enter untuk '{default_name}'): ").strip()

    # Jika kosong → pakai default
    if not out_path:
        out_path = default_name

    # Pastikan ekstensi .csv
    if not out_path.endswith('.csv'):
        out_path += '.csv'

    # Simpan file
    df.to_csv(out_path, index=False)

    success(f"File berhasil disimpan : {C.CYAN}'{out_path}'{C.RESET}")
    info(f"Ukuran akhir           : {df.shape[0]} baris x {df.shape[1]} kolom")
    info(f"Missing value tersisa  : {df.isnull().sum().sum()}")

    return out_path

# ═══════════════════════════════════════════════════════════════
# MAIN
# ═══════════════════════════════════════════════════════════════

def main():
    os.system('cls' if os.name == 'nt' else 'clear')

    print(f"""
{C.CYAN}{C.BOLD}╔══════════════════════════════════════════════════════════════╗
║      PREPROCESSING PIPELINE — FUZZY C-MEANS CLUSTERING      ║
║      COVID-19 & E-Commerce Eropa · Generasi X & Y           ║
╚══════════════════════════════════════════════════════════════╝{C.RESET}
""")

    # Step 1: Muat data hardcoded
    df = load_data()

    # Step 2: Pilihan preprocessing
    do_preprocess = ask_preprocessing(df)

    if not do_preprocess:
        header("SELESAI — TANPA PREPROCESSING")
        warning("Data digunakan dalam kondisi mentah (masih ada missing & outlier).")
        show_table(df, "Data final (mentah):")
        save_result(df)

    else:
        # Step 3: Handle missing values
        df = handle_missing(df)

        # Step 4: Deteksi & koreksi outlier
        df = handle_outlier(df)
        # Step 6: Simpan

        print(df)
        out_path = save_result(df)

        # Ringkasan akhir
        header("PREPROCESSING SELESAI")
        success("Semua tahap preprocessing berhasil dijalankan!")
        print()
        info("Ringkasan pipeline yang dijalankan:")
        print(f"    {C.CYAN}[3]{C.RESET} Handle Missing Values       {C.GREEN}✔ selesai{C.RESET}")
        print(f"    {C.CYAN}[4]{C.RESET} Deteksi & Koreksi Outlier   {C.GREEN}✔ selesai{C.RESET}")
        print(f"    {C.CYAN}[5]{C.RESET} Normalisasi / Standarisasi  {C.GREEN}✔ selesai{C.RESET}")
        print()
        success("Data siap digunakan untuk Fuzzy C-means clustering.")
        info(f"File output : {C.CYAN}{out_path}{C.RESET}")

    print(f"\n{C.DIM}{'─'*62}{C.RESET}\n")


if __name__ == "__main__":
    main()
```

```python
import math
import random
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
class Preproccesing:
    def read_data_csv(self):
        df = pd.read_csv(
            './data-plm.csv', # file csv yang tertera
            delimiter=',',
            encoding='utf-8',
        )

        json_data = df.to_json(orient='records', indent=2)
        json_data = json.loads(json_data)
        result = {}

        for items in json_data:
            val = [int(values) for keys, values in items.items() if keys != 'Country']

            result[items['Country']] = val

        self.data = result
```

```python
class Utilitas(Preproccesing):
    def __init__(self):
        super().__init__()
        self.read_data_csv()

    # convert dari Object country data ke Object per fitur
    def to_object_fitur(self):
        keys = ["All_Individuals_2018", "All_Individuals_2021",	"GenY_2018",	"GenY_2021",
        "GenX_2018",	"GenX_2021",	"Men_GenY_2018",	"Men_GenY_2021",
        "Men_GenX_2018", "Men_GenX_2021", "Women_GenY_2018", "Women_GenY_2021",
        "Women_GenX_2018", "Women_GenX_2021"]

        len_fitur = len(list(self.data.values())[0])
        values = [[] for _ in range(len_fitur)]
        object_val = list(self.data.values())

        for i in range(len(object_val)):
            for j in range(len(object_val[i])):
                values[j].append(object_val[i][j])

        result = {}
        for i, k in enumerate(keys):
            result[k] = values[i]

        return result

    def filter_year(self, data):
        YEAR_COLUMNS = {
            2018: [0, 2, 4, 6, 8, 10, 12],   # 7 fitur tahun 2018
            2021: [1, 3, 5, 7, 9, 11, 13],   # 7 fitur tahun 2021
        }

        if self.year is None:
            return data
        cols = YEAR_COLUMNS[self.year]
        return [[row[i] for i in cols] for row in data]
```

```python
class FCM(Utilitas):
    """
        All_Individuals_2018, All_Individuals_2021,	GenY_2018,	GenY_2021,
        GenX_2018,	GenX_2021,	Men_GenY_2018,	Men_GenY_2021,
        Men_GenX_2018, Men_GenX_2021, Women_GenY_2018, Women_GenY_2021,
        Women_GenX_2018, Women_GenX_2021 (14 features - berurutan)
    """

    def __init__(self, C, m, Year=None):
        super().__init__()
        self.read_data_csv()

        self.total_fitur = len(list(self.data.values()))
        self.total_country = len(list(self.data.keys()))
        self.C = C
        self.m = m
        self.year = Year

    def normalisation_data(self):
        data_country = list(self.data.values())

        if self.year is not None:
            data_country = self.filter_year(data_country)

        data_index = [[] for _ in range(len(data_country[0]))]

        for i in range(len(data_country)):
            for j in range(len(data_country[i])):
                data_index[j].append(data_country[i][j])

        min_value = [min(data_index[i]) for i in range(len(data_index))]
        max_value = [max(data_index[i]) for i in range(len(data_index))]

        data_norm = []
        for i in range(len(data_country)):
            data_i = []

            for j in range(len(data_country[i])):
                x_norm = (data_country[i][j] - min_value[j]) / (max_value[j] - min_value[j])
                data_i.append(round(x_norm, 4))

            data_norm.append(data_i)

        list_country = list(self.data.keys())

        result = {}
        for keys, country in enumerate(list_country):
            result[country] = data_norm[keys]

        return result

    def init_membership(self, n):
        random.seed(42)

        U = []

        for _ in range(n):
            row = [random.random() for _ in range(self.C)]
            total = sum(row)

            row = [round(val / total, 2) for val in row]
            U.append(row)

        return U

    def centroid_calculation(self, U):
        # initial_bobot = self.init_membership(n=self.total_country)
        norm_data = list(self.normalisation_data().values())
        exp = 2 / (self.m - 1)

        sum_cluster = [0 for _ in range(self.C)]
        Cluster = [[] for _ in range(self.C)] # cluster setelah dikuadrat

        for bobot in U:
            cluster = [round(cluster ** exp, 4) for cluster in bobot]

            for keys, c in enumerate(cluster):
                Cluster[keys].append(c)
                sum_cluster[keys] += c

        result_cluster = [
            [
                round(
                    sum(Cluster[i][j] * norm_data[j][k] for j in range(len(norm_data))) / sum_cluster[i],
                    4
                )
                for k in range(len(norm_data[0]))
            ]
            for i in range(len(Cluster))
        ]

        return result_cluster


    def update_membership(self, U):
        list_centroid = self.centroid_calculation(U)
        norm_data = list(self.normalisation_data().values())

        distance = [
            [
                round(math.sqrt(sum(
                    (norm_data[j][k] - list_centroid[i][k]) ** 2
                    for k in range(len(norm_data[0]))
                )), 4)
                for j in range(len(norm_data))
            ]
            for i in range(len(list_centroid))
        ]

        new_U = []
        for j in range(len(norm_data)):  # tiap data
            row = []

            for i in range(len(list_centroid)):  # tiap cluster
                if distance[i][j] == 0:
                    val = 1
                else:
                    denom = 0
                    for k in range(len(list_centroid)):
                        if distance[k][j] == 0:
                            continue
                        ratio = distance[i][j] / distance[k][j]
                        denom += ratio ** (2 / (self.m - 1))

                    val = 1 / denom if denom != 0 else 0

                row.append(round(val, 4))

            new_U.append(row)

        return new_U

    def objective_function(self, U, list_centroid):
        norm_data = list(self.normalisation_data().values())

        J = 0
        for i in range(self.C):
            for j in range(len(norm_data)):
                d_ij = math.sqrt(sum(
                    (norm_data[j][k] - list_centroid[i][k]) ** 2
                    for k in range(len(norm_data[0]))
                ))
                J += (U[j][i] ** self.m) * (d_ij ** 2)

        return round(J, 6)

    def fit(self, max_iter=100, epsilon=1e-4):
        norm_data = list(self.normalisation_data().values())
        n = len(norm_data)

        U = self.init_membership(n)
        history_J = []

        print(f"\nFCM | C={self.C}, m={self.m}, ε={epsilon}")
        print(f"{'─'*45}")
        print(f"{'Iterasi':>8} │ {'Objective Function (J)':>22} │ {'ΔU':>12}")
        print(f"{'─'*45}")

        for iterasi in range(1, max_iter + 1):
            U_lama = [row[:] for row in U]

            centroids = self.centroid_calculation(U)
            U = self.update_membership(U)

            J = self.objective_function(U, centroids)
            history_J.append(J)

            delta = math.sqrt(sum(
                (U[j][i] - U_lama[j][i]) ** 2
                for j in range(n)
                for i in range(self.C)
            ))

            print(f"{iterasi:>8} │ {J:>22.6f} │ {delta:>12.6f}")

            if delta < epsilon:
                print(f"{'─'*45}")
                print(f"Konvergen di iterasi {iterasi}")
                break
        else:
            print(f"{'─'*45}")
            print(f"Belum konvergen setelah {max_iter} iterasi")

        self.U_final   = U
        self.centroids = centroids
        self.history_J = history_J
        self.labels    = [row.index(max(row)) for row in U]

        return self

    # ─────────────────────────────────
    #  VISUALISASI SCATTER PLOT
    # ─────────────────────────────────
    def visualize(self, title="FCM Clustering"):
        import matplotlib.pyplot as plt
        from sklearn.manifold import TSNE
        import numpy as np  # tambah ini

        if not hasattr(self, 'labels'):
            raise RuntimeError("Jalankan .fit() dulu sebelum visualize().")

        norm_data = list(self.normalisation_data().values())
        countries = list(self.data.keys())

        # t-SNE: 14 fitur → 2 dimensi
        # perplexity disesuaikan karena data cuma 35 baris
        norm_array = np.array(norm_data)  # ← tambah ini
        tsne = TSNE(n_components=2, perplexity=4, random_state=42)
        projected = tsne.fit_transform(norm_array)  # shape (35, 2)

        colors = ['steelblue', 'tomato', 'seagreen', 'darkorange',
                'mediumpurple', 'gold', 'deeppink', 'sienna']

        plt.figure(figsize=(11, 7))

        for c in range(self.C):
            idx = [j for j, lbl in enumerate(self.labels) if lbl == c]
            xs  = [projected[j][0] for j in idx]
            ys  = [projected[j][1] for j in idx]

            plt.scatter(xs, ys,
                        color=colors[c % len(colors)],
                        label=f"Cluster {c}",
                        s=100,
                        edgecolors='white',
                        linewidths=0.6,
                        zorder=3)

            # label nama negara
            for j in idx:
                plt.annotate(
                    countries[j],
                    (projected[j][0], projected[j][1]),
                    fontsize=7.5,
                    ha='left', va='bottom',
                    xytext=(5, 5),
                    textcoords='offset points'
                )

        plt.xlabel("t-SNE Dimension 1")
        plt.ylabel("t-SNE Dimension 2")
        plt.title(f"{title} — t-SNE (C={self.C}, m={self.m})")
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.4)
        plt.tight_layout()
        plt.show()
```

```python
def elbow(c_range=range(2, 9), m=2, max_iter=100, epsilon=1e-4):
    J_list = []

    print(f"\n{'─'*40}")
    print(f"{'Elbow Method — Fuzzy C-Means':^40}")
    print(f"{'─'*40}")
    print(f"{'C':>5} │ {'J Akhir':>15} │ {'Iterasi':>8}")
    print(f"{'─'*40}")

    for c in c_range:
        model = FCM(C=c, m=m)
        model.fit(max_iter=max_iter, epsilon=epsilon)
        J_akhir = model.history_J[-1]
        n_iter  = model.n_iter if hasattr(model, 'n_iter') else '-'
        J_list.append(J_akhir)
        # print(f"{c:>5} │ {J_akhir:>15.6f} │ {n_iter:>8}")

    # print(f"{'─'*40}")

    c_values = list(c_range)
    plt.figure(figsize=(8, 5))
    plt.plot(c_values, J_list, marker='o', color='steelblue', linewidth=2)
    plt.xlabel("Jumlah Cluster (C)")
    plt.ylabel("Objective Function (J)")
    plt.title("Elbow Method — Fuzzy C-Means")
    plt.xticks(c_values)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
# semua fitur (14 kolom)
model_all = FCM(C=3, m=2)
model_all.fit()

# khusus 2018 (7 kolom)
model_2018 = FCM(C=3, m=2, Year=2018)
model_2018.fit()

# khusus 2021 (7 kolom)
model_2021 = FCM(C=4, m=2, Year=2021)
model_2021.fit()
```
