```python
# 2.16

# base
s = float(input("Enter the length of side: "))
area = (3 * math.sqrt(3)) / 2 * s**2
print(f"The area of the Hexagon is {area}")

# function
def hexa(sisi):
    area = (3 * math.sqrt(3)) / 2 * sisi**2

    print(f"The area of the Hexagon is {area}")

s = float(input("Enter the length of side: "))
hexa(s)
```

```python
# 2.17

ta = float(input("Enter the temperatur in Fahrenheit betweer -58F and 41 F: "))

while ta < -58 or ta > 41:
    ta = float(input("Enter the temperatur in Fahrenheit betweer -58F and 41 F: "))

v = int(input("Enter the wind speed (>= 2) in miles per hour: "))

while v < 2:
    v = int(input("Enter the wind speed (>= 2) in miles per hour: "))

formula = 35.74 + (0.6215 * ta) - (35.75 * math.pow(v, 0.16)) + (0.4275 * ta * math.pow(v, 0.16))

print(f"The wind chill index is {formula}")
```

```python
# 2.19

print("Enter the coordinates of three points seperated by spaces")

print("like x1 y1 x2 y2 x3 y3: ")

point = []
for i in range(6):
    x = float(input(""))
    point.append(x)

x1, y1, x2, y2, x3, y3 = point

s = (x1 * (y2 - y3) + x2 * (y3 - y1) + (x3 * (y1 - y2)))
area = math.fabs(0.5 * s)

print(f"The area of the triangle is {area:.2f}")
```
